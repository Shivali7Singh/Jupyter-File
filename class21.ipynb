```python
# 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("retail_1data.csv")
df
```

```python
df.describe
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Convert Date column to date time
df["Date"]=pd.to_datetime(df["Date"])
```

```python
# Create month column
df["Month"]=df["Date"].dt.to_period("M")
```

```python
monthly_sales=df.groupby("Month")["Total Amount"].sum()
plt.figure()
monthly_sales.plot(kind="bar")
plt.title("Monthly sales")
plt.xlable("Month")
plt.ylable("Total sales")
plt.show()
```

```python
df
```

```python
cat_sales=df.groupby("Product Category")["Total Amount"].sum()
colors=["green","pink","blue"]
plt.figure(figsize=(4,5)) #Create A new Graph
cat_sales.plot(kind="bar",color=colors)
plt.title("Product Category sales of retail data")
plt.xlable("ProductCategory")
plt.ylable("Total sales")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Quantity"],df["Total Amount"])
plt.title("Quantity vs Total Amount")
plt.xlable("Quantity")
plt.ylable("Total Amount")
plt.show()
```
