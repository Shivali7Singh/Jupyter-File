```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Chocolate_sales_data.csv")
df
```

```python
sales=df.groupby("Sales Person")["Total Amount"].sum()
plt.figure()
sales.plot(kind="bar")
plt.title("Sales Person Total Amount ")
plt.xlable("Sales Person")
plt.ylable("Total Amount")
plt.show()
```
