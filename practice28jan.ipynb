```python
import numpy as np
import pandas as pd
```

```python
# Assuming 'df' is your DataFrame with the given data
df = pd.DataFrame({
    'Student': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Python': [5, 4, np.nan, 2, 3],
    'Data Science': [4, np.nan, 5, 3, np.nan],
    'ML': [np.nan, 5, 4, np.nan, 2],
    'Web Dev': [2, 3, np.nan, 5, 4],
    'Cloud': [np.nan, 2, 3, 4, np.nan]
}).set_index('Student')
```

```python
print(df)
```

```python
# Replace NaN with 0 (or mean/median if needed)
marks_matrix = df.fillna(0).values
```

```python
# Perform SVD
U, S, VT = np.linalg.svd(marks_matrix, full_matrices=False)
```

```python
# Choose k=2
k = 2
U_k = U[:, :k]
S_k = np.diag(S[:k])
VT_k = VT[:k, :]
```

```python
# Reconstruct approximate marks matrix
approx_marks = U_k @ S_k @ VT_k
```

```python
print(approx_marks)
```
